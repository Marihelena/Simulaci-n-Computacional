### Se generan mil valores

```python
print 'n   Xo   (aXo + c) modm   Xn + 1    Numero  '
for i in range (1000): #desde i hasta el numero que se quiere generar
    i += 1
    semilla=7
    mod=16
    xn = ((5*semilla)+3)%mod    #Formula general
    div = xn/mod             #Formula general
    z = float((xn+.0) / (mod+.0))
    print "%d   %d    (%d  +  %d)/%d     %d        %f  "%(i, semilla, div, xn, mod, xn, z)
    semilla=xn           #Xn+1 se convierte en la semilla
```
